##Torres de Hanoi
###Método: Divide y venceras 
(Recursividad y cálculo del numero de operaciónes)
####Torres de Hanoi opera mediante la siguiente ecuación:
$$T(n) = 1+2 * T(N-1) = 1+2*(1+2*T(n-2))= 1+2+2^2+...+2 * T^n(0) = 2^n -1$$
####Complejidad: 
$$ O(2^n)$$




```
Ficha = int(input('¿Cuantas fichas?\n')) ##Solicita el numero de fichas y lo convierte en un valor entero.
def Torres_Hanoi(N, desde, hasta):

  if N==1:
      print('Mueve desde la '+ str(desde) + ' hasta ' + str(hasta))
  else:
     Torres_Hanoi(N-1, desde, 6-hasta-desde)
     print('Mueve desde la '+ str(desde) + ' hasta ' + str(hasta))
     Torres_Hanoi(N-1, 6-hasta-desde, hasta)

Torres_Hanoi(Ficha, 1, 3) ##Llama la función y obtiene el valor agregado en el input.
```

##Sistema Monetario
###Método: Algoritmo Voraz



```
#Estimado para pesos mexicanos
Sistema_Monetario = [1000, 500, 200, 100, 50, 20, 10, 5, 2, 1] #Está adaptado al sistema monetario mexicano

def cambio_monedas(N, SM=Sistema_Monetario):

  Valor_Acumulado = 0

  SOL = [0]*len(SM)

  for i,m in enumerate(SM):
    monedas = int((N-Valor_Acumulado)/m)
    SOL[i] = monedas
    Valor_Acumulado = Valor_Acumulado + monedas*m
    if Valor_Acumulado == N:
      return SOL

print(cambio_monedas(N=int(input('¿Cuanto dinero quieres calcular?\n')), SM= Sistema_Monetario))##Pide de forma abierta el numero de N (Cantidad)
```

###Problema de las Reinas
####Método: Backtracking

```
def reinas(N, solucion, etapa): #Tamaño del tablero

  for i in range(1,N+1):
     solucion[etapa]= i

     #print(solucion)
     if es_prometedora(solucion,etapa):
       if etapa == N-1 :
         print("\n LA solución es:")
         print(solucion)
         #escribe(solucion)
       else:
         print("Es prometedora \n-------------")
         reinas(N,solucion,etapa+1)
     else:
       print("NO PROMETEDORA \n-------------")
       None
      
     solucion[etapa] = 0
```

```
def es_prometedora(SOLUCION, etap):
   #print(SOLUCION)
   #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
   for i in range(etapa+1):
     #print("El valor" + str(SOLUCION[i]) + "está" + str(SOLUCION.count(SOLUCION[i]))+ "veces")
     if SOLUCION.count(SOLUCION[i]) > 1 :      return False 

     #Verifica las diagonales
     for j in range (i+1, etapa+1 ):
       #print("Comprobando diagonal de " + str(i) + "y" + str(j))
       if abs(i-j) == abs(SOLUCION[i] - SOLUCION[j]) : return False
      
   return True
```
